```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
xmlfile = "18001.xml"

tree = ET.parse(xmlfile)
root = tree.getroot()

ET.dump(tree)
```

```python
for elm in root.findall("./"):
    print(elm.text)
```

```python
print(root.tag, root.attrib)
```

```python
for child in root:
    print(child.tag, child.attrib)
```
